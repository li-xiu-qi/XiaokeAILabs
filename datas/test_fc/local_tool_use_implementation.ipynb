## 1. å®šä¹‰å·¥å…·å‡½æ•°å’Œæè¿°

é¦–å…ˆå®šä¹‰æˆ‘ä»¬è¦æä¾›ç»™æ¨¡å‹çš„å·¥å…·å‡½æ•°ã€‚

````python
# å·¥å…·å‡½æ•°å®šä¹‰
def get_current_time() -> str:
    """è·å–å½“å‰æ—¶é—´"""
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def calculate_area(shape: str, **kwargs) -> dict:
    """è®¡ç®—å‡ ä½•å›¾å½¢é¢ç§¯"""
    if shape == "circle":
        radius = kwargs.get("radius", 0)
        if radius <= 0:
            return {"error": "åœ†çš„åŠå¾„å¿…é¡»å¤§äº0"}
        return {"shape": "circle", "area": math.pi * radius ** 2}
    elif shape == "rectangle":
        width = kwargs.get("width", 0)
        height = kwargs.get("height", 0)
        if width <= 0 or height <= 0:
            return {"error": "çŸ©å½¢çš„å®½åº¦å’Œé«˜åº¦å¿…é¡»å¤§äº0"}
        return {"shape": "rectangle", "area": width * height}
    else:
        return {"error": f"ä¸æ”¯æŒçš„å›¾å½¢ç±»å‹: {shape}"}

def get_weather(city: str) -> str:
    """è·å–åŸå¸‚å¤©æ°”ä¿¡æ¯ï¼ˆæ¨¡æ‹Ÿï¼‰"""
    weather_data = {
        "åŒ—äº¬": "æ™´å¤©ï¼Œæ°”æ¸© 15Â°Cï¼Œå¾®é£",
        "ä¸Šæµ·": "å¤šäº‘ï¼Œæ°”æ¸© 18Â°Cï¼Œä¸œå—é£",
        "å¹¿å·": "é›¨å¤©ï¼Œæ°”æ¸© 22Â°Cï¼Œå—é£",
        "æ·±åœ³": "æ™´å¤©ï¼Œæ°”æ¸© 25Â°Cï¼Œæ— é£",
        "æ­å·": "é˜´å¤©ï¼Œæ°”æ¸© 16Â°Cï¼Œè¥¿åŒ—é£"
    }
    return weather_data.get(city, f"æŠ±æ­‰ï¼Œæš‚æ—¶æ— æ³•è·å– {city} çš„å¤©æ°”ä¿¡æ¯")

def search_knowledge(query: str) -> str:
    """æœç´¢çŸ¥è¯†åº“ï¼ˆæ¨¡æ‹Ÿï¼‰"""
    knowledge_base = {
        "python": "Python æ˜¯ä¸€ç§è§£é‡Šå‹ã€é¢å‘å¯¹è±¡ã€åŠ¨æ€æ•°æ®ç±»å‹çš„é«˜çº§ç¨‹åºè®¾è®¡è¯­è¨€ã€‚",
        "æœºå™¨å­¦ä¹ ": "æœºå™¨å­¦ä¹ æ˜¯äººå·¥æ™ºèƒ½çš„ä¸€ä¸ªåˆ†æ”¯ï¼Œè®©è®¡ç®—æœºé€šè¿‡æ•°æ®å­¦ä¹ æ¨¡å¼ã€‚",
        "æ·±åº¦å­¦ä¹ ": "æ·±åº¦å­¦ä¹ æ˜¯æœºå™¨å­¦ä¹ çš„å­é›†ï¼Œä½¿ç”¨ç¥ç»ç½‘ç»œæ¨¡æ‹Ÿäººè„‘å­¦ä¹ è¿‡ç¨‹ã€‚",
        "ai": "äººå·¥æ™ºèƒ½ï¼ˆAIï¼‰æ˜¯è®¡ç®—æœºç§‘å­¦çš„ä¸€ä¸ªé¢†åŸŸï¼Œè‡´åŠ›äºåˆ›å»ºæ™ºèƒ½æœºå™¨ã€‚"
    }
    
    query_lower = query.lower()
    for key, value in knowledge_base.items():
        if key in query_lower:
            return value
    
    return f"æŠ±æ­‰ï¼Œæ²¡æœ‰æ‰¾åˆ°å…³äº '{query}' çš„ç›¸å…³ä¿¡æ¯"

print("âœ… å·¥å…·å‡½æ•°å®šä¹‰å®Œæˆ")
````

## 2. å·¥å…·æè¿°å’Œæ³¨å†Œç³»ç»Ÿ

åˆ›å»ºä¸€ä¸ªå·¥å…·æ³¨å†Œç³»ç»Ÿï¼Œç”¨äºç®¡ç†å·¥å…·çš„æè¿°å’Œè°ƒç”¨ã€‚

````python
@dataclass
class ToolParameter:
    """å·¥å…·å‚æ•°å®šä¹‰"""
    name: str
    type: str
    description: str
    required: bool = True

@dataclass
class ToolDefinition:
    """å·¥å…·å®šä¹‰"""
    name: str
    function: Callable
    description: str
    parameters: List[ToolParameter]

class LocalToolRegistry:
    """æœ¬åœ°å·¥å…·æ³¨å†Œè¡¨"""
    
    def __init__(self):
        self.tools: Dict[str, ToolDefinition] = {}
    
    def register_tool(self, tool_def: ToolDefinition):
        """æ³¨å†Œå·¥å…·"""
        self.tools[tool_def.name] = tool_def
        print(f"âœ… æ³¨å†Œå·¥å…·: {tool_def.name}")
    
    def get_tool(self, name: str) -> Optional[ToolDefinition]:
        """è·å–å·¥å…·å®šä¹‰"""
        return self.tools.get(name)
    
    def list_tools(self) -> List[str]:
        """åˆ—å‡ºæ‰€æœ‰å·¥å…·åç§°"""
        return list(self.tools.keys())
    
    def generate_tool_prompt(self) -> str:
        """ç”Ÿæˆå·¥å…·æè¿°çš„ prompt"""
        if not self.tools:
            return "å½“å‰æ²¡æœ‰å¯ç”¨çš„å·¥å…·ã€‚"
        
        prompt = "ä½ æœ‰ä»¥ä¸‹å·¥å…·å¯ä»¥ä½¿ç”¨ï¼š\n\n"
        
        for tool_name, tool_def in self.tools.items():
            prompt += f"**{tool_name}**: {tool_def.description}\n"
            prompt += "å‚æ•°:\n"
            
            if not tool_def.parameters:
                prompt += "  - æ— éœ€å‚æ•°\n"
            else:
                for param in tool_def.parameters:
                    required_str = "å¿…éœ€" if param.required else "å¯é€‰"
                    prompt += f"  - {param.name} ({param.type}, {required_str}): {param.description}\n"
            prompt += "\n"
        
        prompt += """å½“ä½ éœ€è¦è°ƒç”¨å·¥å…·æ—¶ï¼Œè¯·ä¸¥æ ¼æŒ‰ç…§ä»¥ä¸‹JSONæ ¼å¼è¾“å‡ºï¼š

```json
{
  "tool_call": {
    "name": "å·¥å…·åç§°",
    "arguments": {
      "å‚æ•°å": "å‚æ•°å€¼"
    }
  }
}
```

å¦‚æœä¸éœ€è¦è°ƒç”¨å·¥å…·ï¼Œè¯·ç›´æ¥å›ç­”é—®é¢˜ã€‚"""
        
        return prompt

# åˆ›å»ºå·¥å…·æ³¨å†Œè¡¨å¹¶æ³¨å†Œå·¥å…·
tool_registry = LocalToolRegistry()

# æ³¨å†Œå·¥å…·
tool_registry.register_tool(ToolDefinition(
    name="get_current_time",
    function=get_current_time,
    description="è·å–å½“å‰çš„æ—¥æœŸå’Œæ—¶é—´",
    parameters=[]
))

tool_registry.register_tool(ToolDefinition(
    name="calculate_area",
    function=calculate_area,
    description="è®¡ç®—å‡ ä½•å›¾å½¢çš„é¢ç§¯ï¼ˆæ”¯æŒåœ†å½¢å’ŒçŸ©å½¢ï¼‰",
    parameters=[
        ToolParameter("shape", "string", "å›¾å½¢ç±»å‹ï¼Œå¯é€‰å€¼ï¼šcircleï¼ˆåœ†å½¢ï¼‰ã€rectangleï¼ˆçŸ©å½¢ï¼‰"),
        ToolParameter("radius", "number", "åœ†çš„åŠå¾„ï¼ˆå½“shapeä¸ºcircleæ—¶å¿…éœ€ï¼‰", required=False),
        ToolParameter("width", "number", "çŸ©å½¢çš„å®½åº¦ï¼ˆå½“shapeä¸ºrectangleæ—¶å¿…éœ€ï¼‰", required=False),
        ToolParameter("height", "number", "çŸ©å½¢çš„é«˜åº¦ï¼ˆå½“shapeä¸ºrectangleæ—¶å¿…éœ€ï¼‰", required=False),
    ]
))

tool_registry.register_tool(ToolDefinition(
    name="get_weather",
    function=get_weather,
    description="è·å–æŒ‡å®šåŸå¸‚çš„å¤©æ°”ä¿¡æ¯",
    parameters=[
        ToolParameter("city", "string", "åŸå¸‚åç§°ï¼Œå¦‚ï¼šåŒ—äº¬ã€ä¸Šæµ·ã€å¹¿å·ã€æ·±åœ³ã€æ­å·")
    ]
))

tool_registry.register_tool(ToolDefinition(
    name="search_knowledge",
    function=search_knowledge,
    description="åœ¨çŸ¥è¯†åº“ä¸­æœç´¢ç›¸å…³ä¿¡æ¯",
    parameters=[
        ToolParameter("query", "string", "æœç´¢æŸ¥è¯¢è¯")
    ]
))

print(f"\nğŸ“‹ å·²æ³¨å†Œ {len(tool_registry.tools)} ä¸ªå·¥å…·")
print("å·¥å…·åˆ—è¡¨:", tool_registry.list_tools())
````

## 3. å·¥å…·è°ƒç”¨è§£æå™¨

åˆ›å»ºè§£æå™¨æ¥ä»æ¨¡å‹è¾“å‡ºä¸­æå–å·¥å…·è°ƒç”¨è¯·æ±‚ã€‚

````python
class ToolCallParser:
    """å·¥å…·è°ƒç”¨è§£æå™¨"""
    
    @staticmethod
    def extract_json_from_text(text: str) -> Optional[dict]:
        """ä»æ–‡æœ¬ä¸­æå–JSON"""
        # å°è¯•å¤šç§æ¨¡å¼åŒ¹é…JSON
        patterns = [
            r'```json\s*(\{.*?\})\s*```',  # ```json ... ```
            r'```\s*(\{.*?\})\s*```',      # ``` ... ```
            r'\{[^{}]*"tool_call"[^{}]*\{[^{}]*\}[^{}]*\}',  # ç›´æ¥åŒ¹é…tool_callç»“æ„
        ]
        
        for pattern in patterns:
            matches = re.findall(pattern, text, re.DOTALL | re.IGNORECASE)
            for match in matches:
                try:
                    return json.loads(match)
                except json.JSONDecodeError:
                    continue
        
        # å°è¯•ç›´æ¥è§£ææ•´ä¸ªæ–‡æœ¬
        try:
            return json.loads(text.strip())
        except json.JSONDecodeError:
            pass
        
        return None
    
    @staticmethod
    def parse_tool_call(text: str) -> Optional[dict]:
        """è§£æå·¥å…·è°ƒç”¨"""
        json_data = ToolCallParser.extract_json_from_text(text)
        
        if not json_data:
            return None
        
        # æ£€æŸ¥æ˜¯å¦åŒ…å«tool_callç»“æ„
        if "tool_call" in json_data:
            tool_call = json_data["tool_call"]
            if "name" in tool_call:
                return {
                    "name": tool_call["name"],
                    "arguments": tool_call.get("arguments", {})
                }
        
        # æ£€æŸ¥æ˜¯å¦ç›´æ¥æ˜¯å·¥å…·è°ƒç”¨æ ¼å¼
        if "name" in json_data and "arguments" in json_data:
            return {
                "name": json_data["name"],
                "arguments": json_data.get("arguments", {})
            }
        
        return None

# æµ‹è¯•è§£æå™¨
parser = ToolCallParser()

test_cases = [
    '```json\n{"tool_call": {"name": "get_current_time", "arguments": {}}}\n```',
    '{"tool_call": {"name": "get_weather", "arguments": {"city": "åŒ—äº¬"}}}',
    'æˆ‘éœ€è¦è°ƒç”¨å·¥å…·ï¼š```json\n{"tool_call": {"name": "calculate_area", "arguments": {"shape": "circle", "radius": 5}}}\n```',
    'æ™®é€šå›ç­”ï¼Œä¸éœ€è¦è°ƒç”¨å·¥å…·'
]

print("ğŸ§ª æµ‹è¯•å·¥å…·è°ƒç”¨è§£æå™¨:")
for i, case in enumerate(test_cases, 1):
    result = parser.parse_tool_call(case)
    print(f"æµ‹è¯• {i}: {'âœ… è§£ææˆåŠŸ' if result else 'âŒ æ— å·¥å…·è°ƒç”¨'}")
    if result:
        print(f"   å·¥å…·: {result['name']}, å‚æ•°: {result['arguments']}")
````

## 4. æœ¬åœ°å·¥å…·è°ƒç”¨æ‰§è¡Œå™¨

åˆ›å»ºå·¥å…·è°ƒç”¨çš„æ‰§è¡Œå™¨ã€‚

````python
class LocalToolExecutor:
    """æœ¬åœ°å·¥å…·æ‰§è¡Œå™¨"""
    
    def __init__(self, tool_registry: LocalToolRegistry):
        self.tool_registry = tool_registry
    
    async def execute_tool(self, tool_name: str, arguments: dict) -> dict:
        """æ‰§è¡Œå·¥å…·è°ƒç”¨"""
        try:
            tool_def = self.tool_registry.get_tool(tool_name)
            if not tool_def:
                return {
                    "success": False,
                    "error": f"å·¥å…· '{tool_name}' ä¸å­˜åœ¨",
                    "available_tools": self.tool_registry.list_tools()
                }
            
            # éªŒè¯å¿…éœ€å‚æ•°
            missing_params = []
            for param in tool_def.parameters:
                if param.required and param.name not in arguments:
                    missing_params.append(param.name)
            
            if missing_params:
                return {
                    "success": False,
                    "error": f"ç¼ºå°‘å¿…éœ€å‚æ•°: {', '.join(missing_params)}"
                }
            
            # æ‰§è¡Œå·¥å…·å‡½æ•°
            if arguments:
                result = tool_def.function(**arguments)
            else:
                result = tool_def.function()
            
            return {
                "success": True,
                "tool_name": tool_name,
                "arguments": arguments,
                "result": result
            }
            
        except Exception as e:
            return {
                "success": False,
                "error": f"å·¥å…·æ‰§è¡Œå¤±è´¥: {str(e)}",
                "tool_name": tool_name,
                "arguments": arguments
            }

# åˆ›å»ºæ‰§è¡Œå™¨
executor = LocalToolExecutor(tool_registry)

# æµ‹è¯•æ‰§è¡Œå™¨
print("ğŸ§ª æµ‹è¯•å·¥å…·æ‰§è¡Œå™¨:")

test_executions = [
    ("get_current_time", {}),
    ("calculate_area", {"shape": "circle", "radius": 5}),
    ("get_weather", {"city": "åŒ—äº¬"}),
    ("search_knowledge", {"query": "Python"}),
    ("nonexistent_tool", {})  # æµ‹è¯•ä¸å­˜åœ¨çš„å·¥å…·
]

for tool_name, args in test_executions:
    result = await executor.execute_tool(tool_name, args)
    status = "âœ…" if result["success"] else "âŒ"
    print(f"{status} {tool_name}: {result}")
````

## 5. æœ¬åœ°å·¥å…·è°ƒç”¨ä¸»ç±»

æ•´åˆæ‰€æœ‰ç»„ä»¶ï¼Œåˆ›å»ºå®Œæ•´çš„æœ¬åœ°å·¥å…·è°ƒç”¨ç³»ç»Ÿã€‚

````python
class LocalToolUseChat:
    """æœ¬åœ°å·¥å…·è°ƒç”¨èŠå¤©ç³»ç»Ÿ"""
    
    def __init__(self, api_key: str, base_url: str, model: str, tool_registry: LocalToolRegistry):
        self.client = AsyncFallbackOpenAIClient(
            primary_api_key=api_key,
            primary_base_url=base_url,
            primary_model_name=model
        )
        self.tool_registry = tool_registry
        self.executor = LocalToolExecutor(tool_registry)
        self.parser = ToolCallParser()
    
    def build_system_prompt(self) -> str:
        """æ„å»ºç³»ç»Ÿæç¤ºè¯"""
        return f"""ä½ æ˜¯ä¸€ä¸ªæ™ºèƒ½åŠ©æ‰‹ï¼Œå¯ä»¥ä½¿ç”¨ä»¥ä¸‹å·¥å…·æ¥å¸®åŠ©ç”¨æˆ·ï¼š

{self.tool_registry.generate_tool_prompt()}

è¯·æ³¨æ„ï¼š
1. åªæœ‰åœ¨ç¡®å®éœ€è¦è°ƒç”¨å·¥å…·æ—¶æ‰ä½¿ç”¨å·¥å…·
2. ä¸¥æ ¼æŒ‰ç…§JSONæ ¼å¼è¾“å‡ºå·¥å…·è°ƒç”¨
3. å¦‚æœä¸éœ€è¦å·¥å…·ï¼Œç›´æ¥å›ç­”é—®é¢˜
4. æ¯æ¬¡åªè°ƒç”¨ä¸€ä¸ªå·¥å…·
"""
    
    async def chat(self, user_message: str, conversation_history: List[dict] = None) -> dict:
        """è¿›è¡Œå¯¹è¯ï¼Œæ”¯æŒå·¥å…·è°ƒç”¨"""
        if conversation_history is None:
            conversation_history = []
        
        # æ„å»ºæ¶ˆæ¯
        messages = [
            {"role": "system", "content": self.build_system_prompt()}
        ]
        messages.extend(conversation_history)
        messages.append({"role": "user", "content": user_message})
        
        max_tool_calls = 3  # é˜²æ­¢æ— é™å¾ªç¯
        tool_call_count = 0
        
        while tool_call_count < max_tool_calls:
            try:
                # è°ƒç”¨æ¨¡å‹
                response = await self.client.chat_completions_create(
                    messages=messages,
                    max_tokens=1000,
                    temperature=0.7
                )
                
                assistant_response = response.choices[0].message.content
                
                # å°è¯•è§£æå·¥å…·è°ƒç”¨
                tool_call = self.parser.parse_tool_call(assistant_response)
                
                if tool_call:
                    print(f"ğŸ”§ æ£€æµ‹åˆ°å·¥å…·è°ƒç”¨: {tool_call['name']}")
                    print(f"ğŸ“ å‚æ•°: {tool_call['arguments']}")
                    
                    # æ‰§è¡Œå·¥å…·
                    tool_result = await self.executor.execute_tool(
                        tool_call['name'], 
                        tool_call['arguments']
                    )
                    
                    print(f"âœ… å·¥å…·æ‰§è¡Œç»“æœ: {tool_result}")
                    
                    # å°†å·¥å…·è°ƒç”¨å’Œç»“æœæ·»åŠ åˆ°å¯¹è¯å†å²
                    messages.append({
                        "role": "user", 
                        "content": f"å·¥å…·è°ƒç”¨ç»“æœ: {json.dumps(tool_result, ensure_ascii=False)}"
                    })
                    
                    tool_call_count += 1
                    continue  # ç»§ç»­ä¸‹ä¸€è½®å¯¹è¯
                else:
                    # æ²¡æœ‰å·¥å…·è°ƒç”¨ï¼Œè¿”å›æœ€ç»ˆå“åº”
                    return {
                        "response": assistant_response,
                        "tool_calls_used": tool_call_count,
                        "conversation": messages,
                        "success": True
                    }
                    
            except Exception as e:
                return {
                    "error": f"å¯¹è¯å¤±è´¥: {str(e)}",
                    "tool_calls_used": tool_call_count,
                    "conversation": messages,
                    "success": False
                }
        
        # è¾¾åˆ°æœ€å¤§å·¥å…·è°ƒç”¨æ¬¡æ•°
        return {
            "response": "å·²è¾¾åˆ°æœ€å¤§å·¥å…·è°ƒç”¨æ¬¡æ•°é™åˆ¶",
            "tool_calls_used": tool_call_count,
            "conversation": messages,
            "success": True
        }
    
    async def close(self):
        """å…³é—­å®¢æˆ·ç«¯"""
        await self.client.close()

# åˆ›å»ºæœ¬åœ°å·¥å…·è°ƒç”¨èŠå¤©ç³»ç»Ÿ
local_chat = LocalToolUseChat(
    api_key=DEEPSEEK_API_KEY,
    base_url=DEEPSEEK_BASE_URL,
    model=DEEPSEEK_MODEL,
    tool_registry=tool_registry
)

print("âœ… æœ¬åœ°å·¥å…·è°ƒç”¨èŠå¤©ç³»ç»Ÿåˆå§‹åŒ–å®Œæˆ")
````

````python
# å®šä¹‰æ–°çš„è‡ªå®šä¹‰å·¥å…·
def generate_password(length: int = 12, include_symbols: bool = True) -> str:
    """ç”Ÿæˆéšæœºå¯†ç """
    import random
    import string
    
    if length < 4:
        return "å¯†ç é•¿åº¦è‡³å°‘ä¸º4ä½"
    
    chars = string.ascii_letters + string.digits
    if include_symbols:
        chars += "!@#$%^&*"
    
    password = ''.join(random.choice(chars) for _ in range(length))
    return password

def calculate_bmi(weight: float, height: float) -> dict:
    """è®¡ç®—BMIæŒ‡æ•°"""
    if weight <= 0 or height <= 0:
        return {"error": "ä½“é‡å’Œèº«é«˜å¿…é¡»å¤§äº0"}
    
    bmi = weight / (height ** 2)
    
    if bmi < 18.5:
        category = "åç˜¦"
    elif bmi < 24:
        category = "æ­£å¸¸"
    elif bmi < 28:
        category = "åèƒ–"
    else:
        category = "è‚¥èƒ–"
    
    return {
        "bmi": round(bmi, 2),
        "category": category,
        "weight": weight,
        "height": height
    }

# åˆ›å»ºæ–°çš„å·¥å…·æ³¨å†Œè¡¨å¹¶æ·»åŠ æ‰©å±•å·¥å…·
extended_registry = LocalToolRegistry()

# æ³¨å†ŒåŸæœ‰å·¥å…·
for tool_name, tool_def in tool_registry.tools.items():
    extended_registry.register_tool(tool_def)

# æ³¨å†Œæ–°å·¥å…·
extended_registry.register_tool(ToolDefinition(
    name="generate_password",
    function=generate_password,
    description="ç”Ÿæˆéšæœºå¯†ç ",
    parameters=[
        ToolParameter("length", "number", "å¯†ç é•¿åº¦ï¼Œé»˜è®¤12ä½", required=False),
        ToolParameter("include_symbols", "boolean", "æ˜¯å¦åŒ…å«ç‰¹æ®Šç¬¦å·ï¼Œé»˜è®¤true", required=False)
    ]
))

extended_registry.register_tool(ToolDefinition(
    name="calculate_bmi",
    function=calculate_bmi,
    description="è®¡ç®—BMIæŒ‡æ•°ï¼ˆä½“é‡æŒ‡æ•°ï¼‰",
    parameters=[
        ToolParameter("weight", "number", "ä½“é‡ï¼ˆå…¬æ–¤ï¼‰"),
        ToolParameter("height", "number", "èº«é«˜ï¼ˆç±³ï¼‰")
    ]
))

print(f"âœ… æ‰©å±•å·¥å…·æ³¨å†Œè¡¨åˆ›å»ºå®Œæˆï¼Œå…±æœ‰ {len(extended_registry.tools)} ä¸ªå·¥å…·")

# æµ‹è¯•æ‰©å±•å·¥å…·
async def test_extended_tools():
    """æµ‹è¯•æ‰©å±•å·¥å…·"""
    
    extended_chat = LocalToolUseChat(
        api_key=DEEPSEEK_API_KEY,
        base_url=DEEPSEEK_BASE_URL,
        model=DEEPSEEK_MODEL,
        tool_registry=extended_registry
    )
    
    test_cases = [
        "å¸®æˆ‘ç”Ÿæˆä¸€ä¸ª16ä½çš„å¯†ç ",
        "æˆ‘çš„ä½“é‡æ˜¯70å…¬æ–¤ï¼Œèº«é«˜1.75ç±³ï¼Œå¸®æˆ‘è®¡ç®—BMI",
        "ç”Ÿæˆä¸€ä¸ª8ä½ä¸åŒ…å«ç‰¹æ®Šç¬¦å·çš„å¯†ç "
    ]
    
    print("ğŸ”§ æµ‹è¯•æ‰©å±•å·¥å…·åŠŸèƒ½")
    print("=" * 50)
    
    try:
        for i, question in enumerate(test_cases, 1):
            print(f"\nğŸ“‹ æµ‹è¯• {i}: {question}")
            
            result = await extended_chat.chat(question)
            
            if result["success"]:
                print(f"ğŸ¤– å›å¤: {result['response']}")
                print(f"ğŸ“Š å·¥å…·è°ƒç”¨æ¬¡æ•°: {result['tool_calls_used']}")
            else:
                print(f"âŒ é”™è¯¯: {result.get('error')}")
    
    finally:
        await extended_chat.close()
    
    print("\nâœ… æ‰©å±•å·¥å…·æµ‹è¯•å®Œæˆ")

# è¿è¡Œæ‰©å±•å·¥å…·æµ‹è¯•
await test_extended_tools()
````
